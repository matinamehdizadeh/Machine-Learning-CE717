```python
fig, axes = plt.subplots(2,2, figsize=(8,8))

r1 = np.random.randn(50)/3
r2 = np.random.randn(50)/3
scatter_hue = np.random.choice([True, False], size=50)
r1 += np.ones(50) * ~scatter_hue * 0.5
axes[0][0].set_title('Scatter plot')
sns.scatterplot(x=r1, y=r2, hue=scatter_hue, ax=axes[0][0])

dist_smaples = list(np.random.normal(loc=0.0, scale=1, size=20)) + [8,8,-32]
axes[0][1].set_title('Histogram')
sns.distplot(dist_smaples, bins=20, hist=True, ax=axes[0][1])

pie_values = [10, 30, 2, 45]
axes[1][0].set_title('Pie Chart')
axes[1][0].pie(sorted(pie_values), labels=sorted(pie_values))

box_values = np.concatenate((2 + np.random.randn(20), -5 + 2 * np.random.randn(5)))
axes[1][1].set_title('Box Plot')
sns.boxplot(box_values, ax=axes[1][1]);
```

```python
df['age'][(df['age'] < 18) | (df['age'] > 75)] = None
df['income'][df['income'] > 11] = None
df['share'][df['share'] < 0] = None
df['expenditure'][(df['expenditure'] < 0) | (df['expenditure'] > 2000)] = None
df['months'][df['months'] > 310] = None
df['active'][(df['active'] < 0) | (df['active'] > 35)] = None
```

```python
# Manage missing values here.
for column in df.loc[:,list(df.dtypes != 'object')]:
    df[column][df[column].isnull()] = df[column].median()
for column in df.loc[:,list(df.dtypes == 'object')]:
    df[column][df[column].isnull()] = df[column].value_counts().index[np.argmax(df[column].value_counts())]
```

```python
# Load and show some of data points
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
with np.load('mnist.zip') as data:
    arr, label = data['X'], data['y']
plt.imshow(arr[0])
```

<div dir=rtl>
با دو روش pca و tsne و با استفاده از ۲ تا از مقادیر ویژه ابعاد دیتاست را کاهش دهید.
    برای روش pca مقادیر و بردارهای ویژه‌ی استفاده شده و درصد واریانس حفظ شده را مشخص کنید.
    <br>
    (۵ نمره)

```python
arr = np.reshape(arr, (10000, 784))
pca = PCA(n_components=2)
pca.fit(arr)
arr_pca = pca.transform(arr)

tsne = TSNE(n_components=2, random_state=0)
arr_2d = tsne.fit_transform(arr)
```

```python
print(pca.singular_values_)
print(pca.components_)
print(pca.explained_variance_)
```

<div dir=rtl>
دیتاست‌های حاصل را در قالب نمودار دوبعدی نمایش داده و قابلیت این دو روش را در جداسازی کلاس‌های مختلف و با استفاده از لیبل داده‌ها مقایسه کنید. علت برتری یک روش به روش دیگر را به طور خلاصه توضیح دهید.

    (۴ نمره)

```python
plt.figure(figsize=(10,7))

sns.scatterplot(arr_pca[:, 0], arr_pca[:, 1], hue = label, palette='tab10').set(title='pca')
plt.show()
plt.figure(figsize=(10,7))

sns.scatterplot(arr_2d[:, 0], arr_2d[:, 1], hue = label, palette='tab10').set(title='t-sne')
plt.show()
```

<div dir=rtl>
در روش t-sne داده‌ها بیشتر بر اساس لیبلشان جدا شدند زیرا در روش t-sne احتمال همسایگی داده‌ها را حساب میکنیم در نتیجه اعداد شبیه هم احتمالا در همسایگی هم بودند و از اعداد دیگر فاصله داشتند در نتیجه از هم جدا شدند. این فاصله در pca حفظ نمی‌شود زیرا pca فاصله‌ی اقلیدسی را حساب می‌کند که در فضاهایی با بعد بالا لزوما جواب درستی به ما نمی‌دهد.
</div>

<div dir=rtl>
حال میخواهیم pca را بر روی دیتاست اصلی خود بدون استفاده از کتابخانه‌ی sklearn پیاده سازی کنیم. برای اینکار در ابتدا ستون‌های فیچر را جداکرده و سپس داده‌ها را نرمالایز کنید. لزوم نرمالایز کردن داده‌ها برای روش pca به چه خصوصیتی از این روش باز میگردد؟
    <br>
    (۴ نمره)
</div>

```python
df_norm = df.iloc[0:, :len(df.columns)-1]
df_norm = (df_norm - df_norm.mean()) / (df_norm.max() - df_norm.min())
df_norm.head(10)
```

<div dir=rtl>
در این سوال هم دو ستون کارت و ایدی را حذف می‌کنیم چون اطلاعات اضافه‌ای به ما نمی‌دهد همچنین ستون ثروت نیز حذف می‌شود.
    <br>
pca جهت ماتریس نهایی را با استفاده از ماکسیموم کردن واریانس ماتریس کواریانس به‌دست می‌آود. به عبارتی سعی دارد داده‌ها طوری تصویر کند که از هم پراکنده باشند تا بتوانیم بیشترین اطلاعات را از آنان استخراج کنیم. چون این کار با استفاده از واریانس ماتریس کواریانس انجام می‌شود باید در ابتدا ماتریس را نرمال کنیم تا در بدست آوردن کواریانس به علت تفاوت در اسکیل داده‌ها دچار خطا نشویم.
</div>

<div dir=rtl>
حال ماتریس کوواریانس دیتاست موجود را بدست آورید و مقادیر و بردار‌های ویژه‌ی آن را به ترتیب نزولی مقادیر ویژه نمایش دهید.
    <br>
    (۳ نمره)
</div>

```python
matrix = np.cov(df_norm.T)
eigen = list(np.linalg.eig(matrix))
ind = np.argsort(eigen[0])[::-1]
eigen_new = [eigen[0][ind], eigen[1][ind]]
print('eigen values: '+str(eigen_new[0]))
print('eigen vectors: '+str(eigen_new[1]))
```

<div dir=rtl>
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
    <br>
    (۴ نمره)
</div>

```python
explained_variance=((eigen_new[0]/sum(eigen_new[0]))*100)
plt.figure(figsize=(7,7))

plt.pie(explained_variance, labels=list(df_norm.columns), autopct='%0.4f%%')
plt.show()
```

<div dir=rtl>
با حفظ حداقل ۵۰ درصد واریانس داده‌ها، ماتریس پرتو(projection matrix) را بدست آورید و تبدیل pca را بر روی دیتاست اجرا کنید.
    <br>
    (۵ نمره)
</div>

```python
cum_var = np.cumsum(explained_variance)
index = np.where(cum_var>50)[0][0]
proj_mat = (np.array(eigen_new[1])[:index+1]).T
projected = np.dot(df_norm, proj_mat)
projected
```

<div dir=rtl>
با استفاده از مقادیر متناظر با pcهای اول و دوم هر داده در دیتاست حاصل شده، نموداری رسم کرده و تاثیر این روش را در جداسازی دو کلاس فقیر و ثروتمند تحلیل کنید.
    <br>
    (۴ نمره)
</div>

```python
valW = projected[df['wealth'] == 1]
valP = projected[df['wealth'] == 0]
plt.scatter(valW[:, 0], valW[:, 1], c='green', label='wealthy', alpha = 0.5)
plt.scatter(valP[:, 0], valP[:, 1], c='purple', label='poor', alpha = 0.5)
```
